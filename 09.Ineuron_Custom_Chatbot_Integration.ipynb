```python
from langchain.document_loaders import UnstructuredURLLoader
from langchain.text_splitter import CharacterTextSplitter
from langchain.vectorstores import Pinecone
import pinecone
from langchain.embeddings import HuggingFaceEmbeddings
from transformers import AutoTokenizer, AutoModelForCausalLM
import transformers
from langchain import HuggingFacePipeline
from huggingface_hub import notebook_login
import torch
```

```python
!pip install numpy==1.24.4
```

```python
URLs = ["https://ineuron.ai/"]
```

```python
loader = UnstructuredURLLoader(urls = URLs)
data = loader.load()
```

```python
data
```

## Chunkins

```python
text_splitter=CharacterTextSplitter(separator='\n',
                                    chunk_size=1000,
                                    chunk_overlap=200)
```

```python
text_chunks=text_splitter.split_documents(data)
```

```python
len(text_chunks)
```

```python
text_chunks[0]
```

## Embedding model

```python
embeddings=HuggingFaceEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2')
```

```python
query_result = embeddings.embed_query("Hello world")
len(query_result)
```

```python
import pinecone
# initialize pinecone
pinecone.init(
    api_key=PINECONE_API_KEY,  # find at app.pinecone.io
    environment=PINECONE_API_ENV  # next to api key in console
)
index_name = "ineuronbot"
```

```python
docsearch = Pinecone.from_texts([t.page_content for t in text_chunks], embeddings, index_name=index_name)
```

# Hugging face techniques

```python
notebook_login()
```

```python
tokenizer = AutoTokenizer.from_pretrained("daryl149/llama-2-7b-chat-hf",
                                          use_auth_token=True,)


model = AutoModelForCausalLM.from_pretrained("daryl149/llama-2-7b-chat-hf",
                                             device_map='auto',
                                             torch_dtype=torch.float16,
                                             use_auth_token=True,
                                              load_in_8bit=True,
                                            #  load_in_8bit_fp32_cpu_offload=True
                                              # load_in_4bit=True
                                             )
```

```python
# model="meta-llama/Llama-2-7b-chat-hf"

# this is unofficial
model="daryl149/llama-2-7b-chat-hf"
```

```python
pipe = transformers.pipeline("text-generation",
                model=model,
                tokenizer= tokenizer,
                torch_dtype=torch.bfloat16,
                device_map="auto",
                max_new_tokens = 512,
                do_sample=True,
                top_k=30,
                num_return_sequences=1,
                eos_token_id=tokenizer.eos_token_id
                )
```

```python
llm=HuggingFacePipeline(pipeline=pipe, model_kwargs={'temperature':0})
```

```python
llm.predict("Please provide summary of elon musk")
```

## Initialize the Retrieval QA

```python
from langchain.chains import RetrievalQA
```

```python

qa = RetrievalQA.from_chain_type(llm=llm, chain_type="stuff", retriever=docsearch.as_retriever())
```

```python
query = "Does ineuron have machine learning course?"
```

```python
print(qa.run(query))
```
